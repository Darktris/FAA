```python
%load_ext autoreload
%autoreload 2
%reload_ext autoreload

import numpy as np
from datos import Datos
from estrategiasparticionado.validacion_simple import ValidacionSimple
from preprocesamientoag.preprocesamiento_ag import PreprocesamientoAG
from clasificadores.clasificador_regresion_logistica import ClasificadorRegresionLogistica
import matplotlib.pyplot as plt

dataset=Datos('./conjunto_datos/wdbc.data')
```

```python
probs =np.linspace(0.2,0.8,10)
poblaciones = []
l_variabilidades = []
l_historicos = []

for p in probs:
    ppag = PreprocesamientoAG(tamano_poblacion=10,max_generaciones=10,probabilidad_cruce=p)
    clasificador = ClasificadorRegresionLogistica()

    poblacion, variabilidades, historico_best_fits = ppag.seleccionarAtributos(dataset, clasificador)

    l_variabilidades.append(variabilidades)
    l_historicos.append(historico_best_fits)


    
    generaciones = np.arange(1,len(variabilidades)+1)

for p,variabilidades, historico_best_fits in zip(probs,l_variabilidades,l_historicos):
    
    print("Probabilidad de cruce:",p)
    print("Variabilidad")
    plt.figure()
    plt.plot(generaciones,variabilidades,color = 'red')
    plt.show()
    plt.close()
    
    print("Fitness")
    plt.plot(generaciones,historico_best_fits,color = 'blue')
    plt.show()
    plt.close()
```
